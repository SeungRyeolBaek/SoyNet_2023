### Linear Regression with TensorFlow

#### Hypothesis & Cost

- Hypothesis is individual one-dimensional formula.<br/>
  >$H(x)=wx+b$<br/>

  Lesser cost, better hypothesis.<br/>
 
- Cost Function is diff between real data and estimated data.<br/>
> $cost^i = (H(x)-h)^2 $ <br/>
> $cost(w,b) = \frac{1}{m} \sum ^m _{i=1} (H(x^i)-y^i)^2$<br/>

  Representing how fit the line to our data.

#### Hypothesis & Cost Practice

```python
#building a hypothesis

#X and Y data
x_train = [1, 2, 3]
y_train = [1, 2, 3]

w = tf.Variable(tf.random_normal([1]), name = 'weight') # variable for tensorflow, trainable
b = tf.Variable(tf.random_normal([1]), name = 'bias')

hypothesis = x_train * w + b
```

```python
#cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - y_train)) # 'reduce_mean' for mean
```

```python
#minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost)
```

```python
#launch the graph in a session
sess = tf.Session()

#initializes global variables in the graph
sess.run(tf.global_variables_initializer())
```

```python
#fit the line
for step in range(2001):
  sess.run(train) # training
  if step % 20 == 0:
    print(step, sess.run(cost), sess.run(w), sess.run(b)) # as progress, cost shrink
```

```python
# with placeholders

X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

hypothesis = X * w + b

cost = tf.reduce_mean(tf.square(hypothesis - Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

sess.run(tf.global_variables_initializer())

for step in range(2001):
  costVal, wVal, bVal, _ = sess.run([cost,w,b,train],feed_dict = {X: [1,2,3], Y: [1.1,2.1,3.1]})
  if step % 20 == 0:
    print(step,costVal, wVal, bVal)
```

#### How to Minimize Cost

Gradient Descent Algorithm<br/>
$Cost(w)$를 $w$에 대하여 미분하여 기울기가 0에 수렴하는 부분을 찾는다.<br/>
> $w := w - \alpha \frac{\sigma}{\sigma w}cost(w)$ $(\alpha =$ learning rate$)$<br/>
> $w := w-α\frac{1}{m}\sum^m_{i=1}(wx^i-y^i)x^i$

기울기의 반대방향으로 움직여야 하므로 '빼기'<br/>
$Cost(W,b)$가 Convex Function 인지 확인!

#### How to Minimize Cost Practice
To Simplify, Omit Bias.

```python
import matplotlib.pyplot as plt
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior() # version sync
```

```python
# Cost Function plot

X = [1,2,3]
Y = [1,2,3]

W = tf.placeholder(tf.float32)

hypothesis = X * W

cost = tf.reduce_mean(tf.square(hypothesis - Y))

sess = tf.Session()

sess.run(tf.global_variables_initializer())

WVal = []
costVal = []
for i in range(-30, 50):
  feedW = i * 0.1
  currCost, currW = sess.run([cost,W], feed_dict = {W: feedW})
  WVal.append(currW)
  costVal.append(currCost)

plt.plot(WVal, costVal)
plt.show()
```

```python
#Minimizing Code

xData = [1,2,3]
yData = [1,2,3]

W = tf.Variable(tf.random_normal([1]), name='weight')
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

hypothesis = X * W

cost = tf.reduce_sum(tf.square(hypothesis- Y))

learningRate = 0.1
gradient = tf.reduce_mean((W * X - Y)*X)
descent = W - learningRate * gradient
update = W.assign(descent)

sess = tf.Session()

sess.run(tf.global_variables_initializer())
for step in range(21):
  sess.run(update, feed_dict={X: xData, Y: yData})
  print(step, sess.run(cost, feed_dict={X: xData, Y: yData}), sess.run(W))

#optimizer 써서 쉽게 실행
```
